Exercício 1 — Selecionar colunas específicas

Atividade: Data Frames

Crie um DataFrame com as seguintes colunas: Nome, Idade, Cidade e Profissão.

Mostre na tela apenas as colunas “Nome” e “Profissão”

```python
import pandas as pd
import numpy as np

dados = {
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniela'],
    'idade': [22, 35, 28, 40],
    'cidade': ['Goiânia', 'Anápolis', 'Valparaíso', 'Luziânia'],
    'profissao': ['Engenheira', 'Professor', 'Designer', 'Médico']
}

df = pd.DataFrame(dados)

print(pd.DataFrame(dados))
```

Exercício 2 — Selecionar linhas específicas

Utilize o mesmo DataFrame anterior e exiba apenas as duas primeiras linhas .

```python

print('\n', df[['nome', 'idade']])
```

Exercício 3 — Filtrar dados com condição

Filtre e mostre apenas os registros onde a idade é maior que 30.

```python
print('\n', df.loc[df['idade'] > 30])
```

Exercício 4 — Adicionar nova coluna

Adicione uma nova coluna chamada “Ano de Nascimento”, calculando-a a partir do valor 2025 - Idade.

```python
df['Ano de Nascimento'] = 2025 - df['idade']
print(df)
```

Exercício 5 — Ordenar e atualizar valores

Ordene o DataFrame por idade em ordem crescente.

Atualize o nome de “Bruno” para “Bruno Souza”.

```python
df = df.sort_values(by='idade')
df.loc[df['nome'] == 'Bruno', 'nome'] = 'Bruno Souza'
print(df)
```

Exercício 6 — Ordenação

Atividade: Data Frames

Crie este código no Google Colab:

Resolava as questões abaixo:

1. Ordene por Nota (decrescente)

2.Desempate por Idade (crescente)

3.Reindexe o resultado

4.Exiba o DataFrame final

```python
import pandas as pd

df = pd.DataFrame({
    'aluno': ['Paulo','Ana', 'Bruna','Carlos'],
    'nota': [7.5, 9.0, 8.0, 9.0],
    'idade': [21, 20, 22, 20]
})

print(df)

#Para ordenar os dados
print('\n', df.sort_values(by='nota', ascending=False))

#Para desempate por Idade (crescente)
df_ordenado = df.sort_values(by=['nota', 'idade'], ascending=[False, True])
print('\n', df.sort_values(by=['idade', 'nota'], ascending=[False, True]))

#Para reindexar o resultado
df_ordenado = df_ordenado.reset_index(drop=True)

#Exiba o DataFrame final
print('\n', df_ordenado)
```
