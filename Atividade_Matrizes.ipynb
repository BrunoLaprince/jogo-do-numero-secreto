Exercício 1 – Criação de Matriz
Crie uma matriz 3x5 e preencha manualmente com números inteiros.
Depois, exiba todos os elementos na tela, um por linha.

```python
matriz = []
for i in range(3):
    linha = []
    for j in range(5):
        valor = int(input(f"Digite o elemento [{i}][{j}]: "))
        linha.append(valor)
    matriz.append(linha)

print('Matriz final:')
for linha in matriz:
    print(linha)
```

Exercício 5 – Soma de Duas Matrizes. Crie duas matrizes 4x4 e gere uma terceira matriz contendo a soma dos valores
correspondentes.

```python
import random
matriz1 = []
matriz2 = []
matrizsoma = []
soma = 0
line = int(input('Digite o número de linhas da matriz: '))
column = int(input('Digite o número de colunas da matriz: '))

for i in range(line):
    linha = []
    for j in range(column):
        linha.append(random.randint(1, 101))
    matriz1.append(linha)

for i in range(line):
    linha = []
    for j in range(column):
        linha.append(random.randint(1, 101))
    matriz2.append(linha)

for i in range(line):
    linha = []
    for j in range(column):
        soma = matriz1[i][j] + matriz2[i][j]
        linha.append(soma)
    matrizsoma.append(linha)

print('\n')
for line in matriz1:
    print(line)

print('\n')
for line in matriz2:
    print(line)

print('\n')
for line in matrizsoma:
    print(line)
```

Exercício 6 – Média e Elementos Maiores que a Média

Crie uma matriz 2x3 com valores definidos pelo usuário;

Calcule a média de todos os elementos;

Mostre quais números são maiores que a média.

```python
import random
contador1 = []
contador2 = []
matriz1 = []
matriz2 = []
mediamatriz = []

soma = 0
line = int(input('Digite o número de linhas da matriz: '))
column = int(input('Digite o número de colunas da matriz: '))

for i in range(line):
    linha = []
    for j in range(column):
        linha.append(random.randint(1, 101))
    matriz1.append(linha)

for i in range(line):
    linha = []
    for j in range(column):
        linha.append(random.randint(1, 101))
    matriz2.append(linha)

for i in range(line):
    linha = []
    for j in range(column):
        media = soma(matriz1[i][j]) / soma(matriz2[i][j])
        linha.append(media)
    mediamatriz.append(linha)

print('\n')
for line in matriz1:
    print(line)

print('\n')
for line in matriz2:
    print(line)

print('\n')
for line in mediamatriz:
    print(line)
```

```python
import random
line = int(input('Digite o número de linhas da matriz: '))
column = int(input('Digite o número de colunas da matriz: '))

#Criação da matriz 2x3 com valores aleatórios
matriz = []
for i in range (line):
    line = []
    for j in range (column):
        valor = random.randint(1, 101)
        line.append(valor)
    matriz.append(line)

# Calculando a média de todos os elementos
soma_total = 0
qtd = 0
for linha in matriz:
    for valor in linha:
        soma_total += valor
        qtd += 1
media = soma_total / qtd

#Identificando os elementos maiores que a média
maiores = []
for linha in matriz:
    for valor in linha:
        if valor > media:
            maiores.append(valor)

print('Matriz gerada aleatoriamente:')
for linha in matriz:
    print(linha)

print(f'\nMédia dos elementos: {media}')

if maiores:
    print('Elementos maiores que a média:', maiores)
else:
    print('Não há elementos maiores que a média na matriz.')
```
